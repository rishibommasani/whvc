```python
# IMPORTS
import pandas as pd
import numpy as np
```

```python
# PREPROCESSING THE DATA FOR VISUALIZATION

YOUR_PATH = '/Users/kaylahuang/Desktop/GitHub/whvc/data/'
raw_data_filename = 'raw.csv'

# Add a new row with column totals
df = pd.read_csv(YOUR_PATH + raw_data_filename, index_col=0)
df.loc['Total'] = df.sum(numeric_only=True, axis=0)

# Add a column that sums up values in each row (excluding Indicator and Commitment columns)
df['Commitment Total'] = df.select_dtypes(include=[np.number]).sum(axis=1)

df.loc['Total', 'Indicator'] = '<b>Commitment Subtotal<b>'
df.loc['Total', 'Commitment Total'] = ''

df.to_csv(YOUR_PATH + 'final.csv', index=False)

# show the dataframe
df
```

```python
# CREATING NEW DATA WITH AGGREGATE SCORES

# Calculate percentage by dividing sum by count of rows in each group
df = df.iloc[:-1, :-1] # ignore the total row and the total column
agg_df = df.groupby('Commitment', sort=False).agg(lambda x: x.sum() / len(x) if x.dtype.kind in 'biufc' else x.iloc[0])
agg_df.drop(columns=['Indicator'], inplace=True)

# Calculate row means across all companies
agg_df['Average'] = agg_df.mean(axis=1)

agg_df.to_csv(YOUR_PATH + 'aggregate.csv')
```
